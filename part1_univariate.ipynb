## 4.1 EDA Univariate analisys (According to Research Questions)

### LOCATION: What are central and not central? 
How is the dataset distributed according to central and non-central ZIP codes ?

```python
# Define central zip codes
central_zipcodes = [
    98101, 98104, 98105, 98109, 98121, 98122, 98154, 98164,  # Downtown
    98103, 98107, 98115, 98117,                             # North/Central Seattle
    98118, 98144, 98106, 98108                              # South Seattle (inner)
]

# Create new column 'location_type'
df['location_type'] = df['zipcode'].apply(lambda z: 'central' if z in central_zipcodes else 'non-central')

df.head()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='location_type')
plt.title('Properties: Central vs Non-Central (Outskirts) according to ZIP codes')
plt.xlabel('Location Type')
plt.ylabel('Number of Properties')

# Add value labels on top of bars
for p in ax.patches:
    height = int(p.get_height())
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2, height + 50), 
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## PRICE: Univariate analysis

### Price distribution in DB
1. Select only most recent price for 1st analysis 
2. Sanity check: How many rows dropped? 
3. Create a price per Sqft
4. Plot Price distribution
5. Plot Price distribution for central and non-central ZIPs 

```python
df_latest = df.sort_values('date').drop_duplicates('house_id', keep='last') # Get the latest entry for each house_id

#Sanity check
print(f"Original: {df.shape[0]} rows")
print(f"Latest sales only: {df_latest.shape[0]} rows") 

df_latest.head()

```

### Price per square foot (Most recent sale price)

```python
# creates a new column Price per square foot (Most recent sales)
df_latest['price_per_sqft'] = df_latest['price'] / df_latest['sqft_lot'] # seems as the whole property lot measurement

df_latest.head()

```

### Price per Sqft distribution plot

```python

plt.figure(figsize=(10, 6))
sns.histplot(df_latest['price'], bins=50, kde=True) # bins=50 for better granularity / common default 
plt.title('Distribution of Property Sale Prices (Latest Sale Only)')
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.xlim(0, df_latest['price'].quantile(0.99))  # exclude top 1% outliers
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Undersanding Price distribution in Central Areas (Zip codes)  and Outskirts (non-central)

```python
# Filter for central properties in the latest sales
df_central = df_latest[df_latest['location_type'] == 'central']

sns.histplot(df_central['price'], bins=50, kde=True)
plt.title('Price Distribution in Central ZIP Codes')
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.xlim(0, df_central['price'].quantile(0.99)) # trim top 1% outliers
plt.show()
```

```python
# Filter for non-central (outskirts) properties in the latest sales
df_outskirts = df_latest[df_latest['location_type'] == 'non-central']

# Plot price distribution
sns.histplot(df_outskirts['price'], bins=50, kde=True)
plt.title('Price Distribution in Outskirts (Non-Central ZIP Codes)')
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.xlim(0, df_outskirts['price'].quantile(0.99))  # trim top 1% outliers
plt.show()
```

## Univariate: Seling Price evolution over years and location

```python
# Find properties sold more than once
multiple_sales = df[df.duplicated('house_id', keep=False)].copy()
print(f"Number of properties sold multiple times: {multiple_sales['house_id'].nunique()}")

#OUTPUT = 176
```

## TOP 10 most Sold Properties

```python
# Define central zip codes
central_zipcodes = [98101, 98104, 98105, 98109, 98121, 98122, 98154, 98164, 98103, 98107, 98115, 98117, 98118, 98144, 98106, 98108]    

# Sort multiple_sales properties by date and house_id 
multiple_sales['date'] = pd.to_datetime(multiple_sales['date'])
multiple_sales = multiple_sales.sort_values(by=['house_id', 'date'])

# Select top 5 properties by number of sales
sample_ids = multiple_sales['house_id'].value_counts().head(10).index.tolist() # number of properties to sample
df_sample = multiple_sales[multiple_sales['house_id'].isin(sample_ids)].copy()

# Tag zip code as central or non-central
df_sample['zip_type'] = df_sample['zipcode'].apply(
    lambda z: 'Central' if z in central_zipcodes else 'Non-Central'
)

# Create a combined label column including location type
df_sample['house_label'] = (
    df_sample['house_id'].astype(str) + 
    ' (ZIP ' + df_sample['zipcode'].astype(str) + ', ' + df_sample['zip_type'] + ')'
)

# Plot the evolution of sale prices for top 5 most sold properties
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sample, x='date', y='price', hue='house_label', marker='o')

plt.title('Price Changes Over Time for Top 10 Most Sold Properties (with ZIP Code and Location Type)')
plt.xlabel('Date')
plt.ylabel('Sale Price')
plt.legend(title='House ID (ZIP, Location Type)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

... Seems like outskirts have an active market (more than city center)
